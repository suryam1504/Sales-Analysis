# Question #4 - What products are most often sold together?

```python
df.head()
```

We can see index 2 and 3 have the same Order Id and were also delivered to the same address, that means these 2 products were sold together.

```python
# to see all the rows with duplicated Order ID
df[df['Order ID'].duplicated(keep=False)]   # keep=False considers all similar values as duplicates
```

```python
# Assigning this to a new dataframe duplicate_df
duplicate_df = df[df['Order ID'].duplicated(keep=False)]
```

```python
duplicate_df.head(20)
```

```python
# Getting all the products of a particular Order ID in one line
duplicate_df['Grouped'] = duplicate_df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
```

```python
duplicate_df.head(20)
```

```python
# Dropping duplicated Order ID
duplicate_df = duplicate_df[['Order ID', 'Grouped']].drop_duplicates()
```

```python
duplicate_df.head(20)
```

```python
# Now to count the pairs which occur most 

# Referenced - https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary

from itertools import combinations
from collections import Counter
```

```python
count = Counter()

for row in duplicate_df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
```

```python
count.most_common(10)
```

```python
for key, value in count.most_common(10):
    print(key, value)
```

### Conclusion - Iphone and Lightning Charging Cable were the 2 products sold together the most.

# Plotting the above graph with prices of each product

```python
prices = df.groupby('Product').mean()['Price Each']
```

```python
prices
```

```python
# Referenced - https://stackoverflow.com/questions/14762181/adding-a-y-axis-label-to-secondary-y-axis-in-matplotlib

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g')
ax2.plot(products, prices, 'b-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Sum of Quantities Ordered', color='g')
ax2.set_ylabel('Price of Product', color='b')

ax1.set_xticklabels(products, rotation = 'vertical', size=10)

plt.show()
```
